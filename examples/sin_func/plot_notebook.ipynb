Converting a Jupyter Notebook to Sphinx-Gallery Python
======================================================

Sphinx-Gallery provides a helper script to convert your Jupyter Notebook
into Python with Sphinx-Gallery structure. For example, this page was
originally written as a Jupyter Notebook. The `conf.py` file for the
Sphinx-Gallery documentation uses the helper function to convert it into
Python before the documentation builds, so it is included in the documentation.

As you can see below, math is properly converted:

$$
x \rightarrow \sin(x)
$$ 

Inline math works as well: $x \rightarrow \sin(x)$.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.ion()

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$\sin(x)$')
plt.show()
```
